```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
homogeneous_point = [2, 3, 1]
plt.scatter(homogeneous_point[0], homogeneous_point[1])
plt.grid(visible=True)
plt.legend()
plt.show()
```

```python
homogeneous_line = [2, 3, 5]
x_values = np.linspace(-5,5,50)
y_values = (-homogeneous_line[0] * x_values - homogeneous_line[1]) / homogeneous_line[2]
plt.scatter(x_values, y_values)
plt.show()
```
