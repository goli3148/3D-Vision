```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def rectangle(x=0, y=0, width=1, length=1, step=.1, cetner=False):
    if cetner:
        x = -1*int(width/2)
        y = -1*int(length/2)
    step_inv = int(1 / step)
    vertices = [(x, y), (x + width, y), (x + width, y + length), (x, y + length)]
    # Generate the points along the edges of the square
    points = []
    for i in range(len(vertices)):
        x1, y1 = vertices[i]    
        x2, y2 = vertices[(i + 1) % len(vertices)]
        if x1 == x2:  # Vertical edge        
            points.extend([(x1, y * step, 1) for y in range(min(y1, y2)*step_inv, max(y1, y2)*step_inv + 1, 1)])
        else:  # Horizontal edge        
            points.extend([(x * step, y1, 1) for x in range(min(x1, x2)*step_inv, max(x1, x2)*step_inv + 1, 1)])
    # Separate the x and y coordinates
    x_coords, y_coords, _ = zip(*points)
    return points, x_coords, y_coords
```

```python
def affinity(points, a11, a12, a21, a22, tx=0, ty=0):
    transform_point = []
    iso_mat = np.array([
        [a11,   a12,         tx],
        [a21,   a22,         ty],
        [0,     0,           1]
    ])
    for point in points:
        transform_point.append(iso_mat @ np.array(point).T)
    x_coords, y_coords, _ = zip(*transform_point)
    return transform_point, x_coords, y_coords

def affinity2(points, theta, phi, lambda1, lambda2, tx=0, ty=0):
    D = np.array([
        [lambda1, 0],
        [0, lambda2]
    ])
    A = R(theta) @ R(-phi) @ D @ R(phi)
    aff_mat = np.array([
        [A[0][0], A[0][1], tx],
        [A[1][0], A[1][1], ty],
        [0,       0,        1]
    ])
    transform_point = []
    for point in points:
        transform_point.append(aff_mat @ np.array(point).T)
    x_coords, y_coords, _ = zip(*transform_point)
    return transform_point, x_coords, y_coords
    
def R(angle):
    angle = np.pi * angle / 180
    R = np.array([
        [np.cos(angle), -1*np.sin(angle)],
        [np.sin(angle), np.cos(angle)]
    ])
    return R
```

```python
fig, ax = plt.subplots()


points, x_coords, y_coords = rectangle(x=2, y=3, width=10, length=20, cetner=False)
ax.scatter(x_coords, y_coords, marker='.', color='b')

points2, x_coords2, y_coords2 = affinity(points, a11=2, a12=1, a21=1, a22=3, tx=0, ty=0)
# ax.scatter(x_coords2, y_coords2, marker='.', color='r')

points2, x_coords2, y_coords2 = affinity2(points, theta=0, phi=30, lambda1=1, lambda2=2, tx=0, ty=0)
ax.scatter(x_coords2, y_coords2, marker='.', color='orange')


# Set the axis limits
xlim = 50
ylim = 50
ax.set_xlim(min(x_coords) - xlim, max(x_coords) + xlim)
ax.set_ylim(min(y_coords) - ylim, max(y_coords) + ylim)
# Set the axis labels
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
# Display the grid
ax.grid(True)
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
ax.set_aspect('equal', adjustable='box')
plt.show()
```
